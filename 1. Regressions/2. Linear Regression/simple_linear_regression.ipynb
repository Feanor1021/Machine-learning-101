# Simple Linear Regression

Simple linear regression is a statistical method that summarizes the relationship between 2 quantitative data. It is a method of estimating the dependent variable (Y) based on the values of the independent variables (X). It is assumed that the two variables are linearly related to each other.

Lets consider that we have number of ordered pairs, and let each ordered binary represent another quantity that changes with respect to a quantity. these pairs are scattered in the x-y plane. Simple linear regression is the process of finding the least erroneous linear regression which we can. We can do it with least square method.(https://byjus.com/maths/least-square-method/#:~:text=The%20least%20square%20method%20is,the%20points%20from%20the%20curve.)

In ML we will use Simple Linear Regression if there is a linear correlation in data set. We will split our data into train and test sets. First we will train our model with Simple Linear Regression with train set and then to check the accuracy of the our model we will try it on test set too. 

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("Salary_Data.csv")
x = dataset.iloc[:, :-1].values #Features
y = dataset.iloc[:, -1].values #Dependent values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## Training the Simple Linear Regression model on the Training set

To create the model, we will use an class from scikitlearn, which is LinearRegression and we will instantiate an object. With this object we will create least erroneous linear regression using x_train and y_train (we are training our model with train set).

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train) # we have trained our model with fit method of the LinearRegression class
```

## Predicting the Test set results

After the training step we can begin to predicting. Now our model has trained, and  found the least erroneous regression it can. Using the linear regression which we found we will predict the salary values for test and train sets.

```python
y_Predicted_test = regressor.predict(x_test)
y_Predicted_train = regressor.predict(x_train)
```

## Visualising the Training set results

Just by looking at these 2 sets, we can estimate how well our model is trained or how well it is. But it's not ethereal to see how accurate it is. So we're going to draw a graph with matplotlib.pyplot.

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot( x_train, y_Predicted_train, color = 'blue') #it will give the regression that we will create with the Regression object.
plt.title("Salary vs Experience(Train)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(x_test, y_test, color = "red")
plt.plot(x_train, y_Predicted_train, color = "blue") # we can also give it x_test and y_Predicted_test because y_Predicted_test is created with
# linear regression of our train sets.
plt.title("Salary vs Experince(Test)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Making a single prediction (for example the salary of an employee with 16 years of experience)

```python
exp_employee = [[12]]
print(regressor.predict(exp_employee)) # predict method expects argument of 2d array
```

Therefore, our model predicts that the salary of an employee with 12 years of experience is $ 138967,5.

Important note: Notice that the value of the feature (12 years) was input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting 12 into a double pair of square brackets makes the input exactly a 2D array. Simply put:
